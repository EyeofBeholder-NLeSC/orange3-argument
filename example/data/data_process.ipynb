```python
import pandas as pd

fpath = './data_original.json'
df = pd.read_json(fpath, lines=True)
df
```

```python
# pick the product with the highest number of reviews
df_args = df[['reviewText', 'asin', 'overall']]
df_args = df_args.iloc[df_args.astype(str).drop_duplicates().index]
target_prod = df_args.groupby(by='asin')['reviewText'].count().reset_index(name='count').sort_values(by='count', ascending=False).iloc[0]['asin']
df_args_selected = df_args[df_args['asin'] == target_prod].reset_index(drop=True)
df_args_selected = df_args_selected.drop(columns=['asin'])
df_args_selected
```

```python
# sample a subset from the selection
df_args_sample = df_args_selected.groupby(by='overall').apply(lambda x: x.sample(5)).reset_index(drop=True)
df_args_sample
```

```python
# output data files
df_args_selected.to_json('./data_processed_1prod_full.json', lines=True, orient='records')
df_args_sample.to_json('./data_processed_1prod_sample.json', lines=True, orient='records')
```
